```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from copy import deepcopy
```

# Problem 1

```python
def prob1():
    """
    Create 3 visualizations of the crime data set.
    One of the visualizations should be a histogram.
    The visualizations should be clearly labelled and easy to understand.
    """
    crime_data = pd.read_csv('crime_data.csv', index_col='Year')
    crime_data_cpy = deepcopy(crime_data)

    # Create visualization 1: How is crime divided amongst each category
    crime_data_cpy = deepcopy(crime_data)
    crime_data_cpy.drop(columns=['Population', 'Total', 'Violent', 'Property'], inplace=True)
    crime_data_cpy.plot(title="Which crimes are commited the most?", figsize=(10, 5), ylabel='# of crimes committed')

    # Create visualization 2: How correlated total crime is to population
    crime_data_cpy = deepcopy(crime_data)
    crime_data_cpy[['Violent','Property']].plot(title="Which crimes categories are commited the most?", figsize=(10, 5), ylabel='# of crimes committed')

    # Create visualization 3: Histogram of total crime
    crime_data_cpy = deepcopy(crime_data)
    crime_data_cpy.drop(columns=['Population', 'Total', 'Violent', 'Property'], inplace=True)
    crime_data_cpy.plot.hist(title="Histogram of crime", figsize=(10, 5), xlabel='# of crimes committed')
```

```python
prob1()
```

# Problem 2

```python
def prob2():
    """
    Use crime_data.csv to plot the trends between Larceny and
        1. Violent
        2. Burglary
        3. Aggravated Assault
    Each graph should be clearly labelled and readable.
    One of these variables does not have a linear trend with Larceny.
    Return a string identifying this variable.
    """
    crime_data = pd.read_csv('crime_data.csv', index_col='Year')
    crime_data[['Larceny', 'Violent']].plot.scatter(x='Larceny', y='Violent', title="Larceny vs Violent", figsize=(10, 10))
    crime_data[['Larceny', 'Burglary']].plot.scatter(x='Larceny', y='Burglary', title="Larceny vs Burglary", figsize=(10, 10))
    crime_data[['Larceny', 'Aggravated Assault']].plot.scatter(x='Larceny', y='Aggravated Assault', title="Larceny vs Aggravated Assault", figsize=(10, 10))
    return "Burglary"
```

```python
prob2()
```

# Problem 3

```python
def prob3():
    """
    Use crime_data.csv to display the following distributions.
        1. The distributions of Burglary, Violent, and Vehicle Theft 
            as box plots
        2. The distributions of Vehicle Thefts against Robberies as
            a hexbin plot.
    All plots should be labelled and easy to read.
    """
    crime_data = pd.read_csv('crime_data.csv', index_col='Year')
    crime_data[['Burglary', 'Violent', 'Vehicle Theft']].plot.box(title="Box plot of Burglary, Violent, and Vehicle Theft", figsize=(10, 10), ylabel='# of crimes committed')
    crime_data.plot.hexbin(x='Vehicle Theft', y='Robbery', title="Vehicle Theft vs Robbery", gridsize=20, figsize=(10, 10), xlabel='# of Vehicle Thefts', ylabel='# of Robberies', sharex=False)
```

```python
prob3()
```

```python
def prob4():
    """
    Create 3 plots that compare variables or universities.
    These plots should answer questions about the data, 
    (e.g. What is the distribution of graduation rates? Do schools 
    with lower student to faculty ratios have higher tuition costs? 
    etc.). These plots should be easy to understand and have clear 
    titles, variable names, and citations.
    """
    college_data = pd.read_csv('college.csv', index_col=0)

    # Visualization 1: What is the distribution of graduation rates?
    college_data_cpy = deepcopy(college_data)
    college_data_cpy[['Grad.Rate']].plot.hist(bins=20, xlabel='Grad Rate', ylabel='# of schools', title='Distribution of graduation rates', figsize=(7,7))
    plt.figtext(0.1,-0.02,"SOURCE: James, G., Witten, D., Hastie, T., and Tibshirani, R. (2013)\n \
                An Introduction to Statistical Learning with applications in R,\n \
                https://www.statlearning.com/, Springer-Verlag, New York")

    # Visualization 2: Do schools with lower student to faculty ratios have higher tuition costs?
    college_data_cpy.plot.hexbin(x='S.F.Ratio', y='Outstate', gridsize=40, xlabel='Student to Faculty Ratio', ylabel='Tuition Cost', title='Do schools with lower student to faculty ratios have higher tuition costs?', figsize=(7,7), sharex=False)
    plt.figtext(0.1,-0.02,"SOURCE: James, G., Witten, D., Hastie, T., and Tibshirani, R. (2013)\n \
                An Introduction to Statistical Learning with applications in R,\n \
                https://www.statlearning.com/, Springer-Verlag, New York")

    # Visualization 3: Do schools that have higher graduation rates have more alumni donate?
    college_data_cpy.plot.scatter(x='Grad.Rate', y='perc.alumni', title='Do schools that have higher graduation rates have more alumni donate?', figsize=(7,7), xlabel='Graduation Rate', ylabel='Alumni Donation %')
    plt.figtext(0.1,-0.02,"SOURCE: James, G., Witten, D., Hastie, T., and Tibshirani, R. (2013)\n \
                An Introduction to Statistical Learning with applications in R,\n \
                https://www.statlearning.com/, Springer-Verlag, New York")
```
